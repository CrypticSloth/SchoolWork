## My work to automatically find barcodes in images and croping the image to just show the barcode

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from wand.image import Image
from wand.color import Color
from math import *
```

```python
img = cv2.imread('Christian/test.jpg')
small = cv2.resize(img,(0,0),fx=0.2,fy=0.2)
# plt.imshow(img)
plt.imshow(small)
```

First, detect where there are areas of high contrast (where the sum of one pixel is `cont` different than the sum of the pixel next to it). Then save the coordinates where this high contrast occurs. 

```python
def highContrast(img,cont):
    c = 0
    for y in range(len(img)):
        for x in range(len(img[y])):
            if (abs(sum(img[y][x-1]) - sum(img[y][x]))) > cont:
                yield [y,x]
```

```python
l = list(highContrast(small,275))
```

Then, with the areas of high contrast saved, figure out where the highest density of high contrast areas occurs and this is most likely the barcode in the image. 

```python
y = [l[i][0] for i in range(len(l))]
x = [l[i][1] for i in range(len(l))]
plt.scatter(y,x)
```

We can see in the above plot the outlines of where high contrast are in the image. The block is the barcodes and the circle is that blob of light in the image.

I will then use Kmeans clustering (with k=2) to find the two places that have the most contrasted areas. Then between the two clusters, select the cluster that has the most contrasted areas (or the most points). This cluster will most likely be the barcode.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2).fit(l)
labels = list(kmeans.labels_)

most_points = max(labels,key=labels.count)
most_cont_list = []
for i in range(len(l)):
    if labels[i] == most_points:
        most_cont_list.append(l[i])

# most_cont_list
```

Now with these points, find the four corners for cropping the image. And then crop the image. What is left should be the barcode.

```python
pmin = 10000
pmax = 0
imin = 0
imax = 0
for i in range(len(most_cont_list)):
    if sum(most_cont_list[i]) < pmin:
        pmin = sum(most_cont_list[i])
        imin = i
    if sum(most_cont_list[i]) > pmax:
        pmax = sum(most_cont_list[i])
        imax = i

print(imin)
print(imax)

minpoint = most_cont_list[imin]
maxpoint = most_cont_list[imax]

print(minpoint)
print(maxpoint)
```

```python
img_with_rec = cv2.rectangle(small,(minpoint[1],minpoint[0]),(maxpoint[1],maxpoint[0]),(0,255,0),3)
plt.imshow(img_with_rec)
```

```python
cv2.imwrite("small_book_img_cropped.png",img_with_rec)
```

```python
crop_img = small[minpoint[0]:maxpoint[0],minpoint[1]:maxpoint[1]]
plt.imshow(crop_img)
```

Put this all in one function

```python
def find_crop_barcode(img,cont):
    
    # Discover areas in image with high contrast
    l = list(highContrast(img,cont))
    
    y = [l[i][0] for i in range(len(l))]
    x = [l[i][1] for i in range(len(l))]
    
    # Use KMeans to find two areas with the most high contrast areas
    kmeans = KMeans(n_clusters=2).fit(l)
    labels = list(kmeans.labels_)

    most_points = max(labels,key=labels.count)
    most_cont_list = []
    for i in range(len(l)):
        if labels[i] == most_points:
            most_cont_list.append(l[i])
            
    # Use the points found to crop the image
    pmin = 10000
    pmax = 0
    imin = 0
    imax = 0
    for i in range(len(most_cont_list)):
        if sum(most_cont_list[i]) < pmin:
            pmin = sum(most_cont_list[i])
            imin = i
        if sum(most_cont_list[i]) > pmax:
            pmax = sum(most_cont_list[i])
            imax = i

    minpoint = most_cont_list[imin]
    maxpoint = most_cont_list[imax]

    crop_img = small[minpoint[0]:maxpoint[0],minpoint[1]:maxpoint[1]]

    return(crop_img)
```

And test it

```python
img = cv2.imread('Christian/test.jpg')
small = cv2.resize(img,(0,0),fx=0.2,fy=0.2) # Reduce the size of the image to save on computation time
plt.imshow(small)
```

```python
cropped = find_crop_barcode(small,275)
plt.imshow(cropped)
```

## Summary

We can see that the method I used works well on book barcodes that are larger. It has more trouble getting all of the smaller barcode. This leaves some room for improvement. It also has trouble when the image only has two barcodes. 

In the second image, it picks up on only one of the barcodes (It picks up the smaller one because it is more 'dense' than the larger one I think). Because I used two clusters for the kmeans, it would pick up on only the two separate barcodes, not any noise. So it would choose the cluster with the most points of contrast. This was apparently the smaller one. I would have thought this would be the larger one. Definitely something to look into. 

It is worth trying other methods other than kmeans to sort out dense areas of contrast.

For images with not a lot of noise, the kmeans algorithm will split the barcode itself into two groups. This is good for noisy pictures like the first one, but not for images like the second and third one where the only high contrast areas in the photo are the barcode. 

Different logic for determining if a pixel is in contrast with another should also be looked at. I think I am loosing information by just summing the pixels values and comparing the value of the difference. This comes into issues with blurry pictures, where the contrast is weaker. We can adjust the `cont` value (which is the difference between contrasts of pixels) but this can be finnicky. There is probably a better, more consistant way of calculating contrast.

This work fits into our class by abstracting what an image really is, just a matrix of numbers. Using this information we can apply known techniques like KMeans to distinquish the data that we want (barcodes) from data that we do not (light blobs, other noise).
